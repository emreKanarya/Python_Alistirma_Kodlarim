# GÖREV-5 !!!!!!!!!!!!!
Age değişkenini kategorik değişkene çeviriniz ve agg_df’e ekleyiniz.

• Age sayısal değişkenini kategorik değişkene çeviriniz.
• Aralıkları ikna edici şekilde oluşturunuz.
• Örneğin: ‘0_18', ‘19_23', '24_30', '31_40', '41_70

```python
#Hocanın yaptığı:
agg_df["AGE"].describe().T #Önce describe ile ortalama, max, min vs görelim ve buna göre ayırma işlemini yapalım yaş değişkenimizi.
```

```python
#Age değişkeninin nerelerden bölüneceğini belirtelim:
bins = [0, 18, 23, 30, 40, agg_df["AGE"].max()]
```

```python
#Bölünen noktalara karşılık isimlendirmelerin ne olacağını ifade edelin:
mylabels = ["0_18", "19_23", "24_30", "31_40",  "41_" + str(agg_df["AGE"].max())]
```

```python
#AGE'i bölelim:
agg_df["AGE_CAT"] = pd.cut(agg_df["AGE"], bins, labels = mylabels)
agg_df.head()
```

# GÖREV-6
Yeni seviye tabanlı müşterileri (persona) tanımlayınız.
 
• Yeni seviye tabanlı müşterileri (persona) tanımlayınız ve veri setine değişken olarak ekleyiniz.
• Yeni eklenecek değişkenin adı: customers_level_based
• Önceki soruda elde edeceğiniz çıktıdaki gözlemleri bir araya getirerek "customers_level_based" değişkenini oluşturmanız gerekmektedir.

```python
#Hocanın Yaptığı:
```

```python
#Öncelikle değişken isimlerine bakalım:
agg_df.columns
```

```python
#Gözlem değerlerine nasıl erişiriz:
for row in agg_df.values:
    print(row)
```

```python
#Alt tire("_") ile birleştirmek için şöyle yapıyoruz: 
#Yukarıda bulunan gözlem değerlerinden bize lazım olanlarını seçecek şekilde işlemi gerçekleştirelim.
#Öncelikle burada "country", "source", "sex" ve "age_cat" değişkenlerini kullanacağız. Bunlarda bizim yukarıda sıfırıncı indeks, birinci indeks, ikinci indeks, beşinci indeks değişkenleri bize lazım.
[row[0].upper() + "_" + row[1].upper() + "_" +row[2].upper() + "_" + row[5].upper() for row in agg_df.values]

#Veri setine ekleyelim:
agg_df["CUSTOMERS_LEVEL_BASED"] = [row[0].upper() + "_" + row[1].upper() + "_" +row[2].upper() + "_" + row[5].upper() for row in agg_df.values]
agg_df.head()
```

```python
#Sadece customers_level_based ve price değişkenlerini kullanacağımız için çağıralım ve diğer değişkenleri çıkaralım:
agg_df = agg_df[["CUSTOMERS_LEVEL_BASED", "PRICE"]]
agg_df.head()
```

```python
#Daha sonra "_" yerine "boşluk" eklemek için şöyle yapıyoruz:
for i in agg_df["CUSTOMERS_LEVEL_BASED"].values:
    print(i.split("_"))
```

```python
agg_df.head()
```

```python
#Daha sonra birçok aynı segment olabilir örneğin USA_ANDROID_MALE_0_18 segmentinden birçok sayıda olabilir. Kontrol edelim:
agg_df["CUSTOMERS_LEVEL_BASED"].value_counts() #Bunun ile hangi persona dan kaçar tane var ona bakıyoruz.
```

```python
#Bunları yani yukarıda örneğin 7 tane count u olanı birleştirmek için groupby ile price değişkeninin mean(ortalamasını) alıyoruz:
agg_df = agg_df.groupby("CUSTOMERS_LEVEL_BASED").agg({"PRICE":"mean"})
agg_df
```

```python
#Daha sonra indeks elde etmek için "reset indeks" yapıyoruz:
agg_df =agg_df.reset_index()
agg_df.head()
```

```python
#Daha sonra kontrol edelim çoklayan değerler var mı?
agg_df["CUSTOMERS_LEVEL_BASED"].value_counts() #Çoklayan değer YOK.
```

```python
#Son hali datafraem in:
agg_df.head()
```

```python
agg_df['CUSTOMERS_LEVEL_BASED'] = agg_df[["SOURCE", "SEX", "COUNTRY", 'AGE_CAT']].agg(lambda x: '_'.join(x).upper(), axis=1)
agg_df.head(20)
agg_df[['CUSTOMERS_LEVEL_BASED', "PRICE"]].head()
```
